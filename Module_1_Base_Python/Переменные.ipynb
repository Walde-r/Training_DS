**Переменная — это именованная область памяти** компьютера, адрес которой позволяет получить доступ к данным (это общее определение из программирования, конкретная трактовка которого может меняться от языка к языку).

В самом простом понимании память —  это упорядоченная последовательность ячеек (байт), в которых располагаются данные. Каждая ячейка имеет свой порядковый номер — адрес.

Адреса, они же порядковые номера, сложны для запоминания. Они не являются интуитивно понятными. Во времена зарождения компьютеров программисту необходимо было помнить, в какой участок памяти он сохранил данные. По такому же принципу появились и переменные — для облегчения доступа к данным в памяти.

Ячейки памяти могут иметь разный размер (также его называют ёмкость, число разрядов, длину) в зависимости от устройства конкретной машины. Современные запоминающие устройства обычно имеют размер ячейки памяти равным одной из степеней двойки: 8 бит, 16 бит, 32 бита, 64 бита. В зависимости от типа данных переменной, в область данных переменной может входить несколько ячеек.

В низкоуровневом программировании (на уровне программирования устройств) до сих пор можно встретить работу с ячейками памяти напрямую по их адресам, но в языках высокого уровня, к которым и относится Python, доступ к памяти и данным, которые в них хранятся, происходит через переменные.
