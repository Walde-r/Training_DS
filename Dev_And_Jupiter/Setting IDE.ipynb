✍ Как у профессионального автомеханика есть свой набор инструментов, так и у дата-сайентиста есть набор инструментов, которые помогают ему оптимизировать работу.

Некоторые из этих инструментов стали уже профессиональными стандартами — если вы ими не владеете, вам будет трудно взаимодействовать с коллегами.

Вообще, любой специалист, работающий на Python, «живёт» в четырёх средах:

та, которая позволяет выполнять код (операционная система);
та, в которой создаётся код (IDE);
та, в которой вы презентуете код (например, Jupyter Notebook);
та, в которой вы храните свой код и работаете над проектом совместно с коллегами (например, GitHub).

### Что такое интегрированная среда разработки?

Интегрированная среда разработки, ИСP (англ. Integrated development environment — IDE), также единая среда разработки, ЕСР — комплекс программных средств, используемый программистами для разработки программного обеспечения. 

Среда разработки включает в себя: 

текстовый редактор (позволяет писать код в текстовом виде);
транслятор (позволяет переводить написанный текст в машинный код);
средства автоматизации сборки (позволяют автоматизировать сборку проекта);
отладчик кода (позволяет выполнять программу пошагово, по одной строке кода, и находить в ней ошибки).

При запуске нам доступно наше первое руководство, которое называется «Начало работы с VS Code».

На стартовой странице IDE попросит нас выполнить несколько шагов по её настройке:

- Выбрать цветовую тему. На выбор предлагается три основных вида оформления и возможность скачать дополнительные. Выберите тот, который вам нравится. По умолчанию используется тёмная тема.
- Синхронизировать другие устройства (если VS Code установлен на нескольких устройствах, вы можете объединить их настройки). Можете пропустить этот пункт.
- Ознакомиться с командами. Здесь нам предлагают посмотреть список команд быстрого доступа, чтобы экономить время работы с VS Code. Можно пропустить данный пункт.

Откроем меню Терминал на верхней панели и создадим терминал. На нижней панели программы появится терминал, консоль отладки и другие инструменты (в зависимости от установленных расширений).

**Терминал** — важная часть работы программиста. Он позволяет прописывать команды для операционной системы и облегчать работу с ней.

Для Windows базовая оболочка терминала (командной строки) — PowerShell, для Linux — bash, для MacOS — shell.

Терминал нам понадобится в дальнейшем для выполнения git-команд для работы с системой контроля версий. Также в терминале можно прописывать установку библиотек и ещё много других команд, позволяющих управлять операционной системой.

После того как вы установили расширение для поддержки Python, в нижней строке программы должно появиться указание об окружении.

Если не вдаваться в подробности, то **окружение** — **это изолированная среда** с конкретной версией языка и библиотек.

С его помощью вы можете переключаться между интерпретаторами и созданными виртуальными средами — это позволяет изолировать проекты друг от друга, то есть каждый проект может иметь собственную версию Python и библиотек.

Тема окружений и работа с ними относится к категориям продвинутого управления разработкой проектов. Она довольно сложная, но мы обязательно поговорим о ней далее в курсе в отдельном модуле, посвящённом пакету Anaconda и работе с ним.
